```python
# 累積精度/エントロピー

en_sort_list = []
for j in range(len(orig_list)):
    tmp = orig_list[j][:, orig_list[j][0].argsort()]
    tmp[1] = tmp[1].cumsum()
    for i in range(len(tmp[1])):
        tmp[1][i] = tmp[1][i] / (i + 1)
    en_sort_list.append(tmp)

# for i in range(len(en_sort_list)):
for i in range(4):
    plt.plot(en_sort_list[i][0], en_sort_list[i][1]*100, linewidth=0, marker='o', markersize=1.0, label='%sbit' % (i+1))
plt.legend()
plt.xlabel("entropy")
plt.ylabel("accuracy (%)")
plt.rcParams["font.size"] = 13
save = True
if save:
    plt.savefig("fig/acc-entropy.png")
plt.show()
```

```python
# loss = (1.0 - accuracy)
def loss_bit(entropy):
    loss = np.zeros([len(en_sort_list)])
    for (i, l) in enumerate(en_sort_list):
        loss[i] = 1 - l[1][np.where(l[0] > entropy)[0][0]]
#         print(l[0][np.where(l[0] > entropy)[0][0]])
#     bit = np.arange(len(en_sort_list)) + 1
#     plt.plot(bit, loss, label='entropy=%s' % entropy)
#     plt.show()
    
#     print(loss,entropy)
    return loss
print(len(en_sort_list))
# computation amount
# a = 1/2
# b = 1
a = 1
b = 0
x = (np.arange(len(en_sort_list)) + 1)
comp = a * x + b

# opt_bit = 1
resol =  20  # resolution
en_max = 6  # max entropy
en_list = np.arange(1, resol+1) / resol * 6
scale_factor = 1 / 70

th_en = np.zeros([len(x)])  # list of threshold with entropy
graph = 0  # show all graphs

color = ([('tab:blue', '#1f77b4'),
    ('tab:orange', '#ff7f0e'),
    ('tab:green', '#2ca02c'),
    ('tab:red', '#d62728'),
    ('tab:purple', '#9467bd'),
    ('tab:brown', '#8c564b'),
    ('tab:pink', '#e377c2'),
    ('tab:gray', '#7f7f7f'),
    ('tab:olive', '#bcbd22'),
    ('tab:cyan', '#17becf')])
for i, e in enumerate(en_list):
    loss = loss_bit(e)
    obj_func = loss + comp * scale_factor  # objective function
#     obj_func = comp * loss
    bit = np.arange(len(en_sort_list)) + 1  # x axis for graph
    linestyle = 'dashed' if i % 2 == 1 else 'solid'
    
    if i==0:
        plt.plot(bit, obj_func, label='0<e$\leq$%.1f' % e, color=color[int(i/2)%9][0], linestyle=linestyle, marker='o', markersize=2) 
    else:
        plt.plot(bit, obj_func, label='%.1f<e$\leq$%.1f' % (en_list[i-1], e), color=color[int(i/2)%9][0], linestyle=linestyle, marker='o', markersize=2)
    plt.plot(obj_func.argmin()+1, obj_func.min(), color="black", markersize=5, marker='o') #color='#ee4466'
    if graph:
        print("entropy = %s" % e)
        plt.show()
    th_en[obj_func.argmin()] = e  # update threshold for each bit-width
#     if int(e*10) == 12 or e == 2:
#         print(obj_func)
plt.xlabel("bit-width")
plt.ylabel("objective function")
plt.rcParams["font.size"] = 16
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=5, fontsize=11)
# plt.legend(loc=2, bbox_to_anchor=(-0.2, -0.2), ncol=5, fontsize=11)
plt.savefig("fig/bit_obj_2.pdf", bbox_inches='tight')
plt.show()
print(th_en)  # show threshold
```

```python
# 各ビットごとhigh_en以上のエントロピーのみを対象としてプロット
high_en = 5.5
bit = np.arange(len(orig_list))

high_index = np.zeros([len(en_sort_list)])
for j in range(len(en_sort_list)):
    tmp = en_sort_list[j]
    for i in range(len(tmp[0])):
        if tmp[0][i] > high_en:
            high_index[j] = i
            break
        
for (l, h, b) in zip(orig_list, high_index, bit):
    tmp = l[:, l[0].argsort()]    
    high = -int((len(l[0])-h))    
    en = tmp[1][high:].cumsum()
    for i in range(len(tmp[1][high:])):
        en[i] = en[i] / (i+1)
    plt.plot(tmp[0][high:], en, linewidth=0, marker='o', markersize=1, label='%sbit' % (b + 1))
plt.legend()
plt.ylabel("correct rate")
plt.xlabel("entropy")
plt.show()
```
